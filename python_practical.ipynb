### Data Type Conversion and Re-visualization

Let's convert relevant columns to their appropriate data types and then re-visualize the data type distribution and missing values.

This is a markdown cell serving as a heading, indicating the subsequent steps will involve converting data types and re-visualizing the results.

```python
# Convert 'Start time', 'End time', 'Duration', and 'Team size' to numeric
for col in ['Start time', 'End time', 'Duration', 'Team size']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("Data types after conversion:")
display(df.dtypes)
```

This cell converts specific columns ('Start time', 'End time', 'Duration', 'Team size') to numeric types using `pd.to_numeric`. The `errors='coerce'` argument ensures that any non-numeric values in these columns are converted to `NaN` instead of raising an error. It then prints the updated data types.

```python
# Re-visualize Data Types Distribution
type_counts_after_conversion = df.dtypes.value_counts()

fig = px.pie(
    values=type_counts_after_conversion.values,
    names=type_counts_after_conversion.index.astype(str),
    title="Data Types Distribution After Conversion"
)

fig.show()
```

After the type conversion, this cell re-generates and displays a pie chart of data type distribution. You should now see `float64` types in addition to `object` types.

```python
# Re-visualize Missing Values After Conversion
missing_after_conversion = df.isnull().sum()

fig = px.bar(
    x=missing_after_conversion.index,
    y=missing_after_conversion.values,
    title="Missing Values After Type Conversion",
    labels={"x":"Columns","y":"Missing Count"}
)

fig.show()
```

This cell re-generates a bar chart of missing values after the type conversion. Converting some non-numeric entries to `NaN` might have slightly increased the missing counts for the converted columns.

```python
duplicates = df.duplicated().sum()

fig = go.Figure(go.Indicator(
    mode="number",
    value=duplicates,
    title={"text": "Duplicate Rows"}
))

fig.show()
```

```python
df.isnull().sum()
```

This cell fills any remaining missing values in the DataFrame using the 'forward-fill' method (`ffill`), meaning it propagates the last valid observation forward to next valid observation. The warning about deprecation suggests that in future pandas versions, `df.ffill()` should be used directly.

```python
#After Cleaning Visualization

missing_after = df.isnull().sum()

fig = px.bar(
    x=missing_after.index,
    y=missing_after.values,
    title="Missing Values After Cleaning"
)

fig.show()

#Zoom
#Hover
#Filter
```

### Further Exploratory Data Analysis: Categorical Features and Time Trends

Let's begin by examining the unique values and distributions of our categorical columns. This helps us understand the variety and frequency of different categories.

```python
categorical_cols = df.select_dtypes(include='object').columns

for col in categorical_cols:
    print(f"\n--- {col} ---")
    print(df[col].value_counts())
    fig = px.bar(
        df[col].value_counts().head(10),
        title=f"Top 10 Frequencies of {col}",
        labels={'index': col, 'value': 'Count'}
    )
    fig.show()
```

Now, let's investigate the relationship between these categorical features and the 'Duration'. We can use box plots to see the distribution of 'Duration' for each category, which is excellent for spotting differences in central tendency and spread.

```python
key_categorical_for_duration = ['Activity', 'Shift', 'Contractor', 'Area']

for col in key_categorical_for_duration:
    fig = px.box(
        df,
        x=col,
        y='Duration',
        title=f"Duration Distribution by {col}",
        points='outliers' # Show outliers
    )
    fig.update_layout(xaxis_tickangle=-45)
    fig.show()
```

Finally, let's look at the 'Duration' over time. Although the dataset might be short, a line plot can reveal any daily patterns or trends.

```python
daily_duration = df.groupby('Date')['Duration'].sum().reset_index()

fig = px.line(
    daily_duration,
    x='Date',
    y='Duration',
    title='Total Daily Duration Over Time'
)
fig.update_xaxes(dtick="D1", tickformat="%b %d") # Show every day
fig.show()
```
